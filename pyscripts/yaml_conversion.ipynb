# Convert yaml into json

This files provide some conversion and preprocessing of the files generated by the essentia_streaming_extractor_freesound. The preprocessing and the conversion are needed to use the output files with the gaia library for classification.

```python
import json
import yaml
import os
import glob
```

```python
yamldir='/Volumes/datadisk/hellska/thesis/datasets/annotate_temp/annotation/guitar'
```

```python
ext_ori='yaml'
ext_def='json'
```

```python
files=glob.glob(os.path.join(yamldir, '*.%s' % ext_ori ))
```

```python
for file in files:
    with open(file) as yamlfile:
        yamldata = yaml.load(yamlfile)
        if 'freesound_extractor' in yamldata['metadata']['version']:
            del yamldata['metadata']['version']['freesound_extractor']
        namestr = os.path.basename(file).split(".")[:2]
        outfilename = namestr[0]+'.'+namestr[1]+'.'+ext_def
        txtdata = json.dumps(yamldata)
        with open(os.path.join(yamldir, outfilename), 'wb') as outfile:
            outfile.write(txtdata)
            
```
